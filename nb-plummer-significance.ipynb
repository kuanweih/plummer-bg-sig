```python
%pylab inline
%config InlineBackend.figure_format = 'retina'

import pandas as pd
from src.plummer_bg_sig import *
```

```python
alpha = 0.5    # for plotting data points
```

```python
# df = pd.read_csv('sig_test_s2_0.68_rh.csv').drop(columns=['Unnamed: 0'])
df = pd.read_csv('sig_test_s2_0.5_rh.csv').drop(columns=['Unnamed: 0'])
df.sample(3)
```

```python
sns.set(style="white", color_codes=True, font_scale=1)
fig, ax = plt.subplots(figsize=(18, 6))

region = 'in_dwarf'

sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s2_mean'], ax=ax, color='b')
sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s2_min'], ax=ax, color='b', alpha=alpha)
sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s2_max'], ax=ax, color='b', alpha=alpha)
ax.errorbar(df['GalaxyName'], df[region + '_s2_mean'], yerr=df[region + '_s2_std'], color='b', ls='none')
ax.axhline(2, ls='--', color='b')

sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s5_mean'], ax=ax, color='orange')
sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s5_min'], ax=ax, color='orange', alpha=alpha)
sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s5_max'], ax=ax, color='orange', alpha=alpha)
ax.errorbar(df['GalaxyName'], df[region + '_s5_mean'], yerr=df[region + '_s5_std'], color='orange', ls='none')
ax.axhline(5, ls='--', color='orange')

sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s8_mean'], ax=ax, color='g')
sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s8_min'], ax=ax, color='g', alpha=alpha)
sns.scatterplot(x=df['GalaxyName'], y=df[region + '_s8_max'], ax=ax, color='g', alpha=alpha)
ax.errorbar(df['GalaxyName'], df[region + '_s8_mean'], yerr=df[region + '_s8_std'], color='g', ls='none')
ax.axhline(8, ls='--', color='g')

fig.autofmt_xdate()
ax.set_ylabel('significance')
ax.set_title('significance ' + region)

# plt.savefig('sig_in_dwarf.png', bbox_inches='tight', dpi=300)
```
